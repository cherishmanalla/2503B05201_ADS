#Generation of 2D Data



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import distance

np.random.seed(0)
np.random.rand(1)
```

```python
np.random.seed(0)
data=np.random.multivariate_normal([0,0],[[1,0.5],[0.5,1]],size=500)
print(data.shape)
```

```python
plt.scatter(data[:,0],data[:,1])
plt.xlabel('X1')
plt.ylabel('Y1')
plt.title('2D Multivariate Normal Distribution')
plt.axis('equal')
plt.show()
```

Step 2 : Compute the Mean and Covariance Matrix

```python
mean=np.mean(data,axis=0)
cov=np.cov(data,rowvar=False)
inv_cov=np.linalg.inv(cov)
```

Step 3: Compute Mahalanobis Distance for Each Point

```python
md=[distance.mahalanobis(x,mean,inv_cov) for x in data]
md=np.array(md)
```

Step 4: Visualize the Mahalanobis Distance

```python
plt.figure(figsize=(7,6))
plt.scatter(data[:,0],data[:,1],c=md,cmap='viridis')
plt.colorbar(label='Mahalanobis Distance')
plt.xlabel('X1')
plt.ylabel('Y1')
plt.title('Mahalanobis Distance for Each Point')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

Step 1: Generate Synthetic Data

```python
np.random.seed(0)
X=np.linspace(0,10,30)
y=np.sin(X)+np.random.normal(0,0.2,len(X))
```

Step 2: Train polynomial regression models

```python
degrees=[1,3,9]
plt.figure(figsize=(10,6))

for degree in degrees:
  poly=PolynomialFeatures(degree)
  X_poly=poly.fit_transform(X.reshape(-1,1))
  model=LinearRegression().fit(X_poly,y)
  y_pred=model.predict(X_poly)

  plt.plot(X,y_pred,label=f'Degree {degree}')
```

Step 3 : Visualization

```python
plt.scatter(X,y,color='black', label='Data')
plt.title('Bias Variance Tradeoff')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
np.random.seed(0)
X=np.linspace(0,10,30)
y=np.sin(X)+np.random.normal(0,0.2,len(X))
degrees=[1,3,9]
plt.figure(figsize=(10,6))

for degree in degrees:
  poly=PolynomialFeatures(degree)
  X_poly=poly.fit_transform(X.reshape(-1,1))
  model=LinearRegression().fit(X_poly,y)
  y_pred=model.predict(X_poly)

  plt.plot(X,y_pred,label=f'Degree {degree}')
  plt.scatter(X,y,color='black', label='Data')
plt.title('Bias Variance Tradeoff')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
